# Conceptual Fréchet mean plots

```python
#Modules
import torch

import numpy as np

import os
import pickle
from collections import defaultdict

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting
```

## Helper Functions

```python
# -----------------------------
# 1. Functions for circle plots
# -----------------------------
def frechet_function(theta, data):
    """
    Fréchet function on the circle using squared geodesic distance.
    """
    d = np.angle(np.exp(1j*(theta - data)))
    return np.sum(d**2)

def draw_circle(ax):
    """
    Draw unit circle.
    """
    t = np.linspace(0, 2*np.pi, 400)
    ax.plot(np.cos(t), np.sin(t), color='black', linewidth=1)

def torus(theta, phi, R=3.0, r=1.0):
    """
    Standard torus embedding in R^3.
    """
    X = (R + r*np.cos(phi)) * np.cos(theta)
    Y = (R + r*np.cos(phi)) * np.sin(theta)
    Z = r * np.sin(phi)
    return X, Y, Z

def ang_dist(a, b):
    return np.minimum(np.abs(a - b), 2*np.pi - np.abs(a - b))
```

```python
def load_grid_experiments(folder, models, lams):
    """
    Load pickle files with names:
    {model}_{computation}_{method}_{optimizer}.pkl

    Returns nested dict:
    data[model][computation][method][optimizer]
    """
    # 4-level nested dict
    data = defaultdict(
                lambda: defaultdict(dict)
            )

    for model in models:
        for lam in lams:
                try:
                    fname = f"{model}_{lam}.pkl"
            
                    path = os.path.join(folder, fname)
                    with open(path, "rb") as f:
                        obj = pickle.load(f)
            
                    data[model][lam] = obj
                except:
                    data[model][lam] = None

    return data
```

```python
def plot_normalized_grid_single_model(runtime_data, MODEL, ci=1.0):
    """
    Plot grid_energy for a single model as a 1 x num_lams heatmap row,
    normalized to [0,1] per plot, with a shared colorbar.
    """

    if MODEL not in runtime_data or not runtime_data[MODEL]:
        print(f"{MODEL} has no data, skipping...")
        return

    lams = list(runtime_data[MODEL].keys())
    num_lams = len(lams)

    fig, axes = plt.subplots(
        1, num_lams,
        figsize=(10 * num_lams, 10),
        constrained_layout=True
    )

    if num_lams == 1:
        axes = [axes]

    im = None

    for ax, lam in zip(axes, lams):
        model_data = runtime_data[MODEL][lam]
        if model_data is None:
            ax.axis('off')
            continue

        grid = model_data.get('grid')
        grid_energy = model_data.get('grid_energy')

        if grid is None or grid_energy is None:
            ax.axis('off')
            continue

        if isinstance(grid, torch.Tensor):
            grid = grid.cpu().detach().numpy()
        if isinstance(grid_energy, torch.Tensor):
            grid_energy = grid_energy.cpu().detach().numpy()

        N = grid_energy.shape[0]
        side = int(np.sqrt(N))
        if side * side != N:
            print(f"{MODEL} lam={lam}: Grid size {N} is not a perfect square, skipping...")
            ax.axis('off')
            continue

        energy_grid = grid_energy.reshape((side, side))

        # CI clipping (per plot)
        if ci < 1.0:
            lower = np.percentile(energy_grid, (1 - ci) / 2 * 100)
            upper = np.percentile(energy_grid, (1 + ci) / 2 * 100)
            energy_grid = np.clip(energy_grid, lower, upper)

        # Per-plot normalization
        min_val = energy_grid.min()
        max_val = energy_grid.max()
        if max_val > min_val:
            energy_grid_norm = (energy_grid - min_val) / (max_val - min_val)
        else:
            energy_grid_norm = np.zeros_like(energy_grid)

        x_coords = grid[:, 0].reshape((side, side))
        y_coords = grid[:, 1].reshape((side, side))

        im = ax.imshow(
            energy_grid_norm,
            origin='lower',
            cmap='viridis',
            vmin=0.0,
            vmax=1.0,
            extent=[x_coords.min(), x_coords.max(),
                    y_coords.min(), y_coords.max()]
        )

        ax.set_xticks([])
        ax.set_yticks([])

        lam_str = str(lam).replace("d", ".")
        ax.set_title(rf"$\lambda={lam_str}$", fontsize=50, pad=12)

    # Shared colorbar (properly placed)
    cbar = fig.colorbar(im, ax=axes, shrink=1.0, pad=0.005)
    cbar.ax.tick_params(labelsize=16)

    save_path = f'{figure_path}grid_{MODEL}.pdf'
    fig.savefig(save_path, format='pdf', pad_inches=0.05)
    plt.show()
```

```python
def plot_normalized_grid_single_model(runtime_data, MODEL, ci=1.0):
    """
    Plot:
    (1) First panel: -sum_i ||x_grid - x_data_i||^2
    (2) Remaining panels: stored grid_energy
    All normalized per plot to [0,1], shared colorbar.
    """

    if MODEL not in runtime_data or not runtime_data[MODEL]:
        print(f"{MODEL} has no data, skipping...")
        return

    lams = list(runtime_data[MODEL].keys())
    num_lams = len(lams)

    # +1 for the data-induced energy panel
    fig, axes = plt.subplots(
        1, num_lams + 1,
        figsize=(10 * (num_lams + 1), 10),
        constrained_layout=True
    )

    if num_lams + 1 == 1:
        axes = [axes]

    im = None

    # =============================
    # First panel: data-induced energy
    # =============================

    lam0 = lams[0]
    model_data = runtime_data[MODEL][lam0]

    grid = model_data.get('grid')
    data_sample = model_data.get('data_sample')

    if isinstance(grid, torch.Tensor):
        grid = grid.cpu().detach().numpy()
    if isinstance(data_sample, torch.Tensor):
        data_sample = data_sample.cpu().detach().numpy()

    N = grid.shape[0]
    side = int(np.sqrt(N))
    if side * side != N:
        raise ValueError("Grid size is not a perfect square.")

    # Reshape grid to (d, d, 2)
    grid_reshaped = grid.reshape((side, side, 2))

    # Compute: -sum_i ||x_grid - x_data_i||^2
    # Result shape: (d, d)
    diff = grid_reshaped[:, :, None, :] - data_sample[None, None, :, :]
    energy_data = -np.sum(diff**2, axis=(2, 3))

    # CI clipping
    if ci < 1.0:
        lower = np.percentile(energy_data, (1 - ci) / 2 * 100)
        upper = np.percentile(energy_data, (1 + ci) / 2 * 100)
        energy_data = np.clip(energy_data, lower, upper)

    # Normalize
    min_val = energy_data.min()
    max_val = energy_data.max()
    energy_data_norm = (
        (energy_data - min_val) / (max_val - min_val)
        if max_val > min_val else np.zeros_like(energy_data)
    )

    x_coords = grid[:, 0].reshape((side, side))
    y_coords = grid[:, 1].reshape((side, side))

    im = axes[0].imshow(
        energy_data_norm,
        origin='lower',
        cmap='viridis',
        vmin=0.0,
        vmax=1.0,
        extent=[x_coords.min(), x_coords.max(),
                y_coords.min(), y_coords.max()]
    )

    axes[0].set_xticks([])
    axes[0].set_yticks([])
    axes[0].set_title(r"Fréchet Mean Negative Loss", fontsize=50, pad=12)

    # =============================
    # Remaining panels: stored grid_energy
    # =============================

    for ax, lam in zip(axes[1:], lams):
        model_data = runtime_data[MODEL][lam]
        grid_energy = model_data.get('grid_energy')
        grid = model_data.get('grid')

        if grid_energy is None or grid is None:
            ax.axis('off')
            continue

        if isinstance(grid_energy, torch.Tensor):
            grid_energy = grid_energy.cpu().detach().numpy()
        if isinstance(grid, torch.Tensor):
            grid = grid.cpu().detach().numpy()

        energy_grid = grid_energy.reshape((side, side))

        # CI clipping
        if ci < 1.0:
            lower = np.percentile(energy_grid, (1 - ci) / 2 * 100)
            upper = np.percentile(energy_grid, (1 + ci) / 2 * 100)
            energy_grid = np.clip(energy_grid, lower, upper)

        # Normalize
        min_val = energy_grid.min()
        max_val = energy_grid.max()
        energy_grid_norm = (
            (energy_grid - min_val) / (max_val - min_val)
            if max_val > min_val else np.zeros_like(energy_grid)
        )

        im = ax.imshow(
            energy_grid_norm,
            origin='lower',
            cmap='viridis',
            vmin=0.0,
            vmax=1.0,
            extent=[x_coords.min(), x_coords.max(),
                    y_coords.min(), y_coords.max()]
        )

        ax.set_xticks([])
        ax.set_yticks([])

        lam_str = str(lam).replace("d", ".")
        ax.set_title(rf"$\lambda={lam_str}$", fontsize=50, pad=12)

    # =============================
    # Shared colorbar
    # =============================

    cbar = fig.colorbar(im, ax=axes, shrink=1.0, pad=0.005)
    cbar.ax.tick_params(labelsize=16)

    save_path = f'{figure_path}grid_{MODEL}.pdf'
    fig.savefig(save_path, format='pdf', pad_inches=0.05)
    plt.show()
```

## Plotting Fréchet mean existence and uniqueness

```python
# =============================
# 1. Circle grid
# =============================

theta_grid = np.linspace(0, 2*np.pi, 2000)

# =============================
# 2. Circle: unique Fréchet mean
# =============================

data_unique = np.array([0.2, 0.4, 0.6])
F_unique = np.array([frechet_function(t, data_unique) for t in theta_grid])
mean_theta_unique = theta_grid[np.argmin(F_unique)]

# =============================
# 3. Circle: non-unique Fréchet mean
# =============================

data_nonunique = np.array([0.0, np.pi])
F_nonunique = np.array([frechet_function(t, data_nonunique) for t in theta_grid])
min_val = np.min(F_nonunique)
minimizers = theta_grid[np.isclose(F_nonunique, min_val, atol=1e-8)]

# =============================
# 4. Sphere mesh (S^2)
# =============================

u = np.linspace(0, 2*np.pi, 200)
v = np.linspace(0, np.pi, 200)
U, V = np.meshgrid(u, v)

X_s = np.cos(U) * np.sin(V)
Y_s = np.sin(U) * np.sin(V)
Z_s = np.cos(V)

# Open hemisphere: Z > 0 (unique Fréchet mean region)
unique_mask_s = Z_s > 0
X_s_u, Y_s_u, Z_s_u = X_s.copy(), Y_s.copy(), Z_s.copy()
X_s_u[~unique_mask_s] = np.nan
Y_s_u[~unique_mask_s] = np.nan
Z_s_u[~unique_mask_s] = np.nan

# =============================
# 5. Torus mesh (T^2)
# =============================

R, r = 3.0, 1.0
theta = np.linspace(0, 2*np.pi, 200)
phi   = np.linspace(0, 2*np.pi, 200)
TH, PH = np.meshgrid(theta, phi)

X_t, Y_t, Z_t = torus(TH, PH, R, r)

# Local uniqueness region (convex normal neighborhood)
theta0, phi0, delta = 0.0, 0.0, 1.0

unique_mask_t = (ang_dist(TH, theta0) < delta) & (ang_dist(PH, phi0) < delta)

X_t_u, Y_t_u, Z_t_u = X_t.copy(), Y_t.copy(), Z_t.copy()
X_t_u[~unique_mask_t] = np.nan
Y_t_u[~unique_mask_t] = np.nan
Z_t_u[~unique_mask_t] = np.nan

# =============================
# 6. Figure
# =============================

fig = plt.figure(figsize=(22, 6))

# ---- Circle: unique ----
ax1 = fig.add_subplot(1, 4, 1)
draw_circle(ax1)
ax1.scatter(np.cos(data_unique), np.sin(data_unique),
            color='crimson', s=120, edgecolor='k', zorder=3, label='Data')
ax1.scatter(np.cos(mean_theta_unique), np.sin(mean_theta_unique),
            color='royalblue', s=180, edgecolor='k', zorder=4, label='Fréchet mean')
ax1.set_aspect('equal')
ax1.axis('off')
ax1.legend(frameon=True)
ax1.set_title("Circle: Unique Mean", fontsize=14, fontweight='bold')

# ---- Circle: non-unique ----
ax2 = fig.add_subplot(1, 4, 2)
draw_circle(ax2)
ax2.scatter(np.cos(data_nonunique), np.sin(data_nonunique),
            color='crimson', s=120, edgecolor='k', zorder=3, label='Data')
ax2.scatter(np.cos(minimizers), np.sin(minimizers),
            color='royalblue', s=180, edgecolor='k', zorder=4, label='Fréchet means')
ax2.set_aspect('equal')
ax2.axis('off')
ax2.legend(frameon=True)
ax2.set_title("Circle: Non-Unique Mean", fontsize=14, fontweight='bold')

# ---- Sphere ----
ax3 = fig.add_subplot(1, 4, 3, projection='3d')
ax3.plot_surface(X_s, Y_s, Z_s,
                 color='lightgray', alpha=0.25,
                 rstride=4, cstride=4, linewidth=0)
ax3.plot_surface(X_s_u, Y_s_u, Z_s_u,
                 color='cornflowerblue', alpha=0.9,
                 rstride=4, cstride=4, linewidth=0)

# Center point (north pole)
eps = 0.15
ax3.scatter(0, 0, 1+eps,
            color='green', s=250,
            edgecolor='k', zorder=5)

ax3.view_init(elev=30, azim=45)
ax3.set_box_aspect([1, 1, 1])
ax3.set_axis_off()
ax3.set_title(r"$\mathbb{S}^2$", fontsize=25)

# ---- Torus ----
ax4 = fig.add_subplot(1, 4, 4, projection='3d')
ax4.plot_surface(X_t, Y_t, Z_t,
                 color='lightgray', alpha=0.25,
                 rstride=4, cstride=4, linewidth=0)
ax4.plot_surface(X_t_u, Y_t_u, Z_t_u,
                 color='cornflowerblue', alpha=0.9,
                 rstride=4, cstride=4, linewidth=0)

# Center point (theta0, phi0)
x0, y0, z0 = torus(theta0, phi0, R, r)
eps = 0.15
ax4.scatter(x0, y0, z0,
            color='green', s=250,
            edgecolor='k', zorder=5)

ax4.view_init(elev=65, azim=35)
ax4.set_box_aspect([1, 1, 1])
ax4.set_axis_off()
ax4.set_title(r"$\mathbb{T}^2$", fontsize=25)

#plt.tight_layout()

save_path = f'{figure_path}frechet_exist_and_uniq.pdf'
fig.savefig(save_path, format='pdf', pad_inches=0.05)
plt.show()

plt.show()
```
