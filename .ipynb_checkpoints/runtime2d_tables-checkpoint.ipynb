## Helper Functions

### Loading data

```python
def load_metric_experiments(folder, models, computations, methods, optimizers):
    """
    Load pickle files with names:
    {model}_{computation}_{method}_{optimizer}.pkl

    Returns nested dict:
    data[model][computation][method][optimizer]
    """
    # 4-level nested dict
    data = defaultdict(
        lambda: defaultdict(
            lambda: defaultdict(dict)
        )
    )

    for model in models:
        for computation in computations:
            for method in methods:
                for optimizer in optimizers:
                    try:
                        fname = f"{model}_{computation}_{method}_{optimizer}.pkl"
                
                        path = os.path.join(folder, fname)
                        with open(path, "rb") as f:
                            obj = pickle.load(f)
                
                        data[model][computation][method][optimizer] = obj
                    except:
                        data[model][computation][method][optimizer] = None

    return data
```

```python
def load_runtime_experiments(folder, models, computations, methods, lams, optimizers):
    """
    Load pickle files with names:
    {model}_{computation}_{method}_{optimizer}.pkl

    Returns nested dict:
    data[model][computation][method][optimizer]
    """
    # 4-level nested dict
    data = defaultdict(
        lambda: defaultdict(
            lambda: defaultdict(
                lambda: defaultdict(dict)
            )
        )
    )

    for model in models:
        for computation in computations:
            for method in methods:
                for lam in lams:
                    for optimizer in optimizers:
                        try:
                            fname = f"{model}_{computation}_{method}_{lam}_{optimizer}.pkl"
                    
                            path = os.path.join(folder, fname)
                            with open(path, "rb") as f:
                                obj = pickle.load(f)
                    
                            data[model][computation][method][lam][optimizer] = obj
                        except:
                            data[model][computation][method][lam][optimizer] = None

    return data
```

```python
def load_grid_experiments(folder, models, lams):
    """
    Load pickle files with names:
    {model}_{computation}_{method}_{optimizer}.pkl

    Returns nested dict:
    data[model][computation][method][optimizer]
    """
    # 4-level nested dict
    data = defaultdict(
                lambda: defaultdict(dict)
            )

    for model in models:
        for lam in lams:
                try:
                    fname = f"{model}_{lam}.pkl"
            
                    path = os.path.join(folder, fname)
                    with open(path, "rb") as f:
                        obj = pickle.load(f)
            
                    data[model][lam] = obj
                except:
                    data[model][lam] = None

    return data
```

```python
def load_grid_experiments(folder, models):
    """
    Load pickle files with names:
    {model}_{computation}_{method}_{optimizer}.pkl

    Returns nested dict:
    data[model][computation][method][optimizer]
    """
    # 4-level nested dict
    data = defaultdict(dict)

    for model in models:
        try:
            fname = f"{model}.pkl"
    
            path = os.path.join(folder, fname)
            with open(path, "rb") as f:
                obj = pickle.load(f)
    
            data[model] = obj
        except:
            data[model] = None

    return data
```

### Creating Tables

```python
def escape_latex(s):
    """Escape special LaTeX characters in strings."""
    return (
        s.replace("_", r"\_")
        .replace("&", r"\&")
        .replace("%", r"\%")
        .replace("$", r"\$")
    )

def latex_sci(x, fmt="{:.3f}", sci_fmt="{:.2f}", threshold=1000):
    """
    Format a number for LaTeX.
    Uses scientific notation if |x| >= threshold.
    Returns '-' if x is None, NaN, or not finite.
    """
    if x is None or not isinstance(x, (int, float)) or not math.isfinite(x):
        return "-"
    if abs(x) < threshold:
        return fmt.format(x)  # plain text (no math mode)

    exponent = int(math.floor(math.log10(abs(x))))
    mantissa = x / (10 ** exponent)

    # Wrap scientific notation in math mode
    return (
        rf"${sci_fmt.format(mantissa)} \times 10^{{{exponent}}}$"
    )

def print_latex_table_by_computation(
    data,
    models,
    methods,
    computations,
    optimizer,
    optimizers_for_runtime=None,
    float_fmt="{:.3f}",
    time_fmt="{:.2f}",
    sci_threshold=1000,
):
    """
    Print a fully integrated stacked LaTeX table for multiple computations and a
    runtime-by-optimizer section at the end. Model headers are printed only once.
    Section headers have bottom rules with spacing to avoid touching text.
    """
    n_cols = 1 + 3 * len(models)

    print(r"\resizebox{\textwidth}{!}{%")
    print(r"\begin{tabular}{l" + "ccc" * len(models) + "}")

    # ---------- Table header: models + metrics (printed once) ----------
    print(r"\toprule")
    header1 = [" "]
    for model in models:
        # Make model names large and bold
        header1.append(r"\multicolumn{3}{c}{\bfseries\large " + model + "}")
    print(" & ".join(header1) + r" \\")
    col = 2
    for _ in models:
        print(rf"\cmidrule(lr){{{col}-{col+2}}}", end=" ")
        col += 3
    print()
    header2 = ["Method"]
    for _ in models:
        header2.extend(["Energy", "NLL", "Runtime"])
    print(" & ".join(header2) + r" \\")
    print(r"\midrule")

    # ---------- Main stacked tables ----------
    for c_idx, computation in enumerate(computations):

        # Extra bottomrule + spacing before header if not first table
        if c_idx > 0:
            print(r"\bottomrule")
            print(r"\addlinespace[0.3em]")  # spacing to avoid header touching rule

        # Compute best values per model
        best = {model: {"Energy": math.inf, "NLL": math.inf, "runtime": math.inf} for model in models}
        for model in models:
            for method in methods:
                obj = data[model][computation][method][optimizer]
                if not isinstance(obj, dict):
                    continue
                e_val = obj.get("Energy", None)
                if isinstance(e_val, (int, float)) and math.isfinite(e_val):
                    best[model]["Energy"] = min(best[model]["Energy"], e_val)
                n_val = obj.get("NLL", None)
                if isinstance(n_val, (int, float)) and math.isfinite(n_val):
                    best[model]["NLL"] = min(best[model]["NLL"], n_val)
                mean_v = obj.get("mean_time", None)
                if isinstance(mean_v, torch.Tensor):
                    mean_v = mean_v.item()
                if isinstance(mean_v, (int, float)) and math.isfinite(mean_v):
                    best[model]["runtime"] = min(best[model]["runtime"], mean_v)

        # Computation header with bottom rule and spacing
        print(rf"\multicolumn{{{n_cols}}}{{c}}{{\bfseries\large {computation}}} \\")
        print(r"\bottomrule")
        print(r"\addlinespace[0.3em]")  # spacing after header

        # Body
        for method in methods:
            row = [method]
            for model in models:
                obj = data[model][computation][method][optimizer]
                e_val = n_val = rt_mean = rt_std = None
                if isinstance(obj, dict):
                    v = obj.get("Energy", None)
                    if isinstance(v, (int, float)) and math.isfinite(v):
                        e_val = v
                    v = obj.get("NLL", None)
                    if isinstance(v, (int, float)) and math.isfinite(v):
                        n_val = v
                    mean_v = obj.get("mean_time", None)
                    std_v = obj.get("std_time", None)
                    if isinstance(mean_v, torch.Tensor):
                        mean_v = mean_v.item()
                    if isinstance(std_v, torch.Tensor):
                        std_v = std_v.item()
                    if isinstance(mean_v, (int, float)) and isinstance(std_v, (int, float)):
                        rt_mean = mean_v
                        rt_std = std_v

                e_str = n_str = rt_str = "-"
                if e_val is not None:
                    e_str = latex_sci(e_val, fmt=float_fmt, threshold=sci_threshold)
                    if e_val == best[model]["Energy"]:
                        e_str = rf"\pmb{{{e_str}}}" if abs(e_val) >= sci_threshold else r"\textbf{" + e_str + "}"
                if n_val is not None:
                    n_str = latex_sci(n_val, fmt=float_fmt, threshold=sci_threshold)
                    if n_val == best[model]["NLL"]:
                        n_str = rf"\pmb{{{n_str}}}" if abs(n_val) >= sci_threshold else r"\textbf{" + n_str + "}"
                if rt_mean is not None and rt_std is not None:
                    mean_str = latex_sci(rt_mean, fmt=time_fmt, threshold=sci_threshold)
                    std_str = latex_sci(rt_std, fmt=time_fmt, threshold=sci_threshold)
                    rt_str = f"{mean_str} $\\pm$ {std_str}"
                    if rt_mean == best[model]["runtime"]:
                        rt_str = rf"\pmb{{{rt_str}}}" if abs(rt_mean) >= sci_threshold else r"\textbf{" + rt_str + "}"
                row.extend([e_str, n_str, rt_str])
            print(" & ".join(row) + r" \\")
        print(r"\addlinespace[0.5em]")

    # ---------- Integrated "Runtime" section ----------
    if optimizers_for_runtime:
        method = "bvp"
        computation = "ProbGEORCE"

        if computations:
            print(r"\bottomrule")
            print(r"\addlinespace[0.3em]")

        print(rf"\multicolumn{{{n_cols}}}{{c}}{{\bfseries\large Runtime}} \\")
        print(r"\bottomrule")
        print(r"\addlinespace[0.3em]")

        # Compute best per model (Energy, NLL, runtime)
        best = {model: {"Energy": math.inf, "NLL": math.inf, "runtime": math.inf} for model in models}
        for model in models:
            for opt in optimizers_for_runtime:
                method_data = data[model][method][computation].get(opt, None)
                if isinstance(method_data, dict):
                    obj = method_data
                else:
                    obj = {}
                # Energy
                e_val = obj.get("Energy", None) if isinstance(obj, dict) else None
                if isinstance(e_val, (int, float)) and math.isfinite(e_val):
                    best[model]["Energy"] = min(best[model]["Energy"], e_val)
                # NLL
                n_val = obj.get("NLL", None) if isinstance(obj, dict) else None
                if isinstance(n_val, (int, float)) and math.isfinite(n_val):
                    best[model]["NLL"] = min(best[model]["NLL"], n_val)
                # Runtime
                mean_v = obj.get("mean_time", None) if isinstance(obj, dict) else None
                if isinstance(mean_v, torch.Tensor):
                    mean_v = mean_v.item()
                if isinstance(mean_v, (int, float)) and math.isfinite(mean_v):
                    best[model]["runtime"] = min(best[model]["runtime"], mean_v)

        # Body
        for opt in optimizers_for_runtime:
            row = [opt]
            for model in models:
                method_data = data[model][method][computation].get(opt, None)
                if isinstance(method_data, dict):
                    obj = method_data
                else:
                    obj = {}
                if isinstance(obj, dict):
                    e_val = obj.get("Energy", None)
                    n_val = obj.get("NLL", None)
                    mean_v = obj.get("mean_time", None)
                    std_v = obj.get("std_time", None)
                else:
                    e_val = n_val = mean_v = std_v = None
                if isinstance(mean_v, torch.Tensor):
                    mean_v = mean_v.item()
                if isinstance(std_v, torch.Tensor):
                    std_v = std_v.item()

                e_str = n_str = rt_str = "-"
                if isinstance(e_val, (int, float)) and math.isfinite(e_val):
                    e_str = latex_sci(e_val, fmt=float_fmt, threshold=sci_threshold)
                    if e_val == best[model]["Energy"]:
                        e_str = rf"\pmb{{{e_str}}}" if abs(e_val) >= sci_threshold else r"\textbf{" + e_str + "}"
                if isinstance(n_val, (int, float)) and math.isfinite(n_val):
                    n_str = latex_sci(n_val, fmt=float_fmt, threshold=sci_threshold)
                    if n_val == best[model]["NLL"]:
                        n_str = rf"\pmb{{{n_str}}}" if abs(n_val) >= sci_threshold else r"\textbf{" + n_str + "}"
                if isinstance(mean_v, (int, float)) and isinstance(std_v, (int, float)):
                    mean_str = latex_sci(mean_v, fmt=time_fmt, threshold=sci_threshold)
                    std_str = latex_sci(std_v, fmt=time_fmt, threshold=sci_threshold)
                    rt_str = f"{mean_str} $\\pm$ {std_str}"
                    if mean_v == best[model]["runtime"]:
                        rt_str = rf"\pmb{{{rt_str}}}" if abs(mean_v) >= sci_threshold else r"\textbf{" + rt_str + "}"
                row.extend([e_str, n_str, rt_str])
            print(" & ".join(row) + r" \\")

    print(r"\bottomrule")
    print(r"\end{tabular}}")
```

```python
def print_latex_table_by_lambda_bvp(
    data,
    models,
    optimizers,
    lams,
    float_fmt="{:.3f}",
    time_fmt="{:.2f}",
    sci_threshold=1000,
):
    """
    Print a LaTeX table where each subtable corresponds to a λ (lambda) value.
    Rows are optimizers, columns are models, metrics: Energy, NLL, Runtime.
    Computation is fixed to 'bvp', matching the plots.
    Adds a horizontal line above each λ section (except the first one) for separation.
    """
    n_cols = 1 + 3 * len(models)

    print(r"\resizebox{\textwidth}{!}{%")
    print(r"\begin{tabular}{l" + "ccc" * len(models) + "}")

    # ---------- Table header: models + metrics ----------
    print(r"\toprule")
    header1 = [" "]
    for model in models:
        header1.append(r"\multicolumn{3}{c}{\bfseries\large " + model + "}")
    print(" & ".join(header1) + r" \\")
    col = 2
    for _ in models:
        print(rf"\cmidrule(lr){{{col}-{col+2}}}", end=" ")
        col += 3
    print()
    header2 = ["Optimizer"]
    for _ in models:
        header2.extend(["Energy", "NLL", "Runtime"])
    print(" & ".join(header2) + r" \\")
    print(r"\midrule")

    # ---------- Loop over λ values ----------
    for i, lam in enumerate(lams):
        # Add a separating horizontal line before all λ headers except the first
        if i > 0:
            print(r"\midrule")

        # Header for each λ
        print(rf"\multicolumn{{{n_cols}}}{{c}}{{\bfseries\large $\lambda = {lam}$}} \\")
        print(r"\bottomrule")
        print(r"\addlinespace[0.3em]")

        # Compute best per model for this lambda across optimizers
        best = {model: {"Energy": math.inf, "NLL": math.inf, "runtime": math.inf} for model in models}
        for model in models:
            for opt in optimizers:
                obj = data.get(model, {}).get('bvp', {}).get('ProbGEORCE', {}).get(lam, {}).get(opt, {})
                if not isinstance(obj, dict):
                    continue
                # Energy
                e_val = obj.get("Energy", None)
                if isinstance(e_val, (int, float)) and math.isfinite(e_val):
                    best[model]["Energy"] = min(best[model]["Energy"], e_val)
                # NLL
                n_val = obj.get("NLL", None)
                if isinstance(n_val, (int, float)) and math.isfinite(n_val):
                    best[model]["NLL"] = min(best[model]["NLL"], n_val)
                # Runtime
                mean_v = obj.get("mean_time", None)
                if isinstance(mean_v, torch.Tensor):
                    mean_v = mean_v.item()
                if isinstance(mean_v, (int, float)) and math.isfinite(mean_v):
                    best[model]["runtime"] = min(best[model]["runtime"], mean_v)

        # ---------- Rows: optimizers ----------
        for opt in optimizers:
            row = [opt]
            for model in models:
                obj = data.get(model, {}).get('bvp', {}).get('ProbGEORCE', {}).get(lam, {}).get(opt, {})
                if not isinstance(obj, dict):
                    obj = {}

                # Energy, NLL
                e_val = obj.get("Energy", None)
                n_val = obj.get("NLL", None)

                # Runtime
                mean_v = obj.get("mean_time", None)
                std_v = obj.get("std_time", None)
                if isinstance(mean_v, torch.Tensor):
                    mean_v = mean_v.item()
                if isinstance(std_v, torch.Tensor):
                    std_v = std_v.item()

                e_str = n_str = rt_str = "-"
                if isinstance(e_val, (int, float)) and math.isfinite(e_val):
                    e_str = latex_sci(e_val, fmt=float_fmt, threshold=sci_threshold)
                    if e_val == best[model]["Energy"]:
                        e_str = rf"\pmb{{{e_str}}}" if abs(e_val) >= sci_threshold else r"\textbf{" + e_str + "}"
                if isinstance(n_val, (int, float)) and math.isfinite(n_val):
                    n_str = latex_sci(n_val, fmt=float_fmt, threshold=sci_threshold)
                    if n_val == best[model]["NLL"]:
                        n_str = rf"\pmb{{{n_str}}}" if abs(n_val) >= sci_threshold else r"\textbf{" + n_str + "}"
                if isinstance(mean_v, (int, float)) and isinstance(std_v, (int, float)):
                    mean_str = latex_sci(mean_v, fmt=time_fmt, threshold=sci_threshold)
                    std_str = latex_sci(std_v, fmt=time_fmt, threshold=sci_threshold)
                    rt_str = f"{mean_str} $\\pm$ {std_str}"
                    if mean_v == best[model]["runtime"]:
                        rt_str = rf"\pmb{{{rt_str}}}" if abs(mean_v) >= sci_threshold else r"\textbf{" + rt_str + "}"
                row.extend([e_str, n_str, rt_str])

            print(" & ".join(row) + r" \\")
        print(r"\addlinespace[0.5em]")

    print(r"\bottomrule")
    print(r"\end{tabular}}")
```

### Plotting

```python
def plot_normalized_grid(runtime_data, models, ci=0.95):
    """
    Plot grid_energy for each model as a 1 x len(lams) heatmap row, normalized to [0, 1] after optional clipping.
    
    Parameters:
        runtime_data: dict containing MODEL -> {lam_value -> {'grid', 'grid_energy'}}
        models: list of model names
        ci: float in (0,1], percentile for clipping (0.95 = 95% CI)
            - If ci=1, no clipping
    """
    import matplotlib.pyplot as plt
    import numpy as np
    import torch
    import os

    for MODEL in models:
        # Check if model exists
        if MODEL not in runtime_data or not runtime_data[MODEL]:
            print(f"{MODEL} has no data, skipping...")
            continue

        lams = runtime_data[MODEL].keys()
        num_lams = len(lams)

        fig, axes = plt.subplots(1, num_lams, figsize=(5*num_lams, 4))
        if num_lams == 1:
            axes = [axes]  # Ensure it's iterable

        for ax, lam in zip(axes, lams):
            model_data = runtime_data[MODEL][lam]
            if model_data is None or 'grid' not in model_data or 'grid_energy' not in model_data:
                print(f"{MODEL} lam={lam} missing data, skipping...")
                ax.axis('off')
                continue

            grid = model_data['grid']
            grid_energy = model_data['grid_energy']

            # Convert to NumPy if tensors
            if isinstance(grid, torch.Tensor):
                grid = grid.cpu().detach().numpy()
            if isinstance(grid_energy, torch.Tensor):
                grid_energy = grid_energy.cpu().detach().numpy()

            # Determine grid size
            N = grid.shape[0]
            side = int(np.sqrt(N))
            if side * side != N:
                print(f"{MODEL} lam={lam}: Grid size {N} is not a perfect square, skipping...")
                ax.axis('off')
                continue

            # Reshape
            energy_grid = grid_energy.reshape((side, side))

            # Clip based on CI if ci < 1
            if ci < 1.0:
                lower = np.percentile(energy_grid, (1-ci)/2*100)
                upper = np.percentile(energy_grid, (1 + ci)/2*100)
                energy_grid = np.clip(energy_grid, lower, upper)

            # Normalize to [0,1]
            min_val = np.min(energy_grid)
            max_val = np.max(energy_grid)
            if max_val > min_val:
                energy_grid_norm = (energy_grid - min_val) / (max_val - min_val)
            else:
                energy_grid_norm = np.zeros_like(energy_grid)

            # Reshape coordinates
            x_coords = grid[:, 0].reshape((side, side))
            y_coords = grid[:, 1].reshape((side, side))

            # Plot
            c = ax.imshow(energy_grid_norm, origin='lower', aspect='auto', cmap='viridis',
                          extent=[x_coords.min(), x_coords.max(), y_coords.min(), y_coords.max()])
            ax.set_title(f"lam={lam}")
            ax.set_xlabel("X")
            ax.set_ylabel("Y")
            fig.colorbar(c, ax=ax, label=f"Normalized Energy")

        fig.suptitle(MODEL)
        save_path = f'{figure_path}grid_{MODEL}.pdf'
        fig.savefig(save_path, format='pdf', pad_inches=0.1, bbox_inches='tight')
        plt.show()
```

### Plotting runtime

```python
FOCUS_METHOD = "ProbGEORCE_Adaptive"
PAD_FACTOR_ENERGY = 2.0   # Padding for Reg Energy zoom
PAD_FACTOR_TIME = 2.0     # Padding for Runtime zoom

fig, axes = plt.subplots(
    2, len(models),
    figsize=(4 * len(models), 8),
    sharey=False
)

# Ensure axes is always 2D array
if len(models) == 1:
    axes = axes.reshape(2, 1)

# =========================
# --- Top row: Reg Energy
# =========================
for col, model in enumerate(models):
    ax = axes[0, col]
    focus_values = []

    for method in optimizers:
        xs, ys = [], []

        for lam in lams:
            try:
                entry = runtime_data.get(model)
                if entry is None:
                    continue

                entry = entry.get('bvp')
                if entry is None:
                    continue

                entry = entry.get('ProbGEORCE')
                if entry is None:
                    continue

                entry = entry.get(lam)
                if entry is None:
                    continue

                entry = entry.get(method)
                if entry is None:
                    continue

                reg_energy = entry.get('NLL')  # or 'Reg Energy'
                if reg_energy is None:
                    continue

                value = reg_energy

                # --- NaN / inf guard ---
                if not np.isfinite(value):
                    continue

                xs.append(lam)
                ys.append(value)

                if method == FOCUS_METHOD:
                    focus_values.append(value)

            except (AttributeError, TypeError):
                continue

        if xs:
            if method == FOCUS_METHOD:
                ax.plot(
                    xs, ys, marker='o',
                    linewidth=3, markersize=7,
                    color='black', label=method, zorder=10
                )
            else:
                ax.plot(
                    xs, ys, marker='o',
                    linewidth=1.2, alpha=0.6,
                    label=method
                )

    # --- Zoom y-axis around focus method (NaN-safe) ---
    focus_values = [v for v in focus_values if np.isfinite(v)]
    if focus_values:
        ymin = min(focus_values)
        ymax = max(focus_values)
        pad = PAD_FACTOR_ENERGY * (
            ymax - ymin if ymax > ymin else max(abs(ymax), 1e-6)
        )
        ax.set_ylim(ymin - pad, ymax + pad)

    ax.set_title(model)

    if col == 0:
        ax.set_ylabel("Reg Energy")

    ax.set_xticklabels([])
    ax.grid(True)

# =========================
# --- Bottom row: Runtime
# =========================
for col, model in enumerate(models):
    ax = axes[1, col]
    adaptive_means = []

    for method in optimizers:
        xs, means, cis = [], [], []

        for lam in lams:
            try:
                entry = runtime_data.get(model)
                if entry is None:
                    continue

                entry = entry.get('bvp')
                if entry is None:
                    continue

                entry = entry.get('ProbGEORCE')
                if entry is None:
                    continue

                entry = entry.get(lam)
                if entry is None:
                    continue

                entry = entry.get(method)
                if entry is None:
                    continue

                mean_time = entry.get('mean_time')
                std_time = entry.get('std_time')
                if mean_time is None or std_time is None:
                    continue

                mean_val = mean_time.cpu().detach().item()
                std_val = std_time.cpu().detach().item()

                # --- NaN / inf guard ---
                if not (np.isfinite(mean_val) and np.isfinite(std_val)):
                    continue

                ci_95 = 1.96 * std_val

                xs.append(lam)
                means.append(mean_val)
                cis.append(ci_95)

                if method == FOCUS_METHOD:
                    adaptive_means.append(mean_val)

            except (AttributeError, TypeError):
                continue

        if not xs:
            continue

        xs = np.array(xs)
        means = np.array(means)
        cis = np.array(cis)

        if method == FOCUS_METHOD:
            ax.plot(xs, means, color='black', linewidth=3, label=method, zorder=10)
            ax.fill_between(xs, means - cis, means + cis,
                            color='black', alpha=0.2, zorder=5)
        else:
            ax.plot(xs, means, linewidth=1.2, alpha=0.6, label=method)
            ax.fill_between(xs, means - cis, means + cis, alpha=0.1)

    # --- Zoom y-axis around focus method (NaN-safe) ---
    adaptive_means = [v for v in adaptive_means if np.isfinite(v)]
    if adaptive_means:
        ymin = min(adaptive_means)
        ymax = max(adaptive_means)
        pad = PAD_FACTOR_TIME * (
            ymax - ymin if ymax > ymin else max(abs(ymax), 1e-6)
        )
        ax.set_ylim(ymin - pad, ymax + pad)

    if col == 0:
        ax.set_ylabel("Runtime")

    ax.set_xlabel("λ (lambda)")
    ax.grid(True)

# =========================
# --- Global legend
# =========================
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc="upper center", ncol=len(labels))

fig.tight_layout()

save_path = f'{figure_path}runtime_comparison.pdf'
fig.savefig(save_path, format='pdf', pad_inches=0.1, bbox_inches='tight')
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Example: pull from your runtime_data
grid = runtime_data['ebm']['0d0']['grid']          # shape: (N_grid, 2)
data_sample = runtime_data['ebm']['0d0']['data_sample']  # shape: (N_data, 2)

# Ensure both are torch tensors
if not isinstance(grid, torch.Tensor):
    grid = torch.tensor(grid)
if not isinstance(data_sample, torch.Tensor):
    data_sample = torch.tensor(data_sample)

# Compute squared distance sum for each grid point
# E(x_grid) = sum_i ||x_grid - x_i||^2
diff = grid.unsqueeze(1) - data_sample.unsqueeze(0)   # shape: (N_grid, N_data, 2)
squared_dist = (diff**2).sum(dim=2)                   # shape: (N_grid, N_data)
energy = squared_dist.sum(dim=1)                      # shape: (N_grid,)

# Reshape to square grid
N = grid.shape[0]
side = int(np.sqrt(N))
if side * side != N:
    raise ValueError("Grid is not a perfect square")
energy_grid = energy.reshape(side, side)

# Reshape coordinates for axes
x_coords = grid[:, 0].reshape(side, side)
y_coords = grid[:, 1].reshape(side, side)

# Compute mean of the data points
mean_point = data_sample.mean(dim=0)  # shape: (2,)

# Move to CPU if needed for plotting
mean_x = mean_point[0].cpu().item()
mean_y = mean_point[1].cpu().item()

# Plot heatmap
plt.figure(figsize=(5,4))
plt.imshow(energy_grid.cpu().detach().numpy(), 
           origin='lower', aspect='auto', cmap='viridis',
           extent=[x_coords.min().cpu(), x_coords.max().cpu(), 
                   y_coords.min().cpu(), y_coords.max().cpu()])

# Overlay mean point
plt.scatter(mean_x, mean_y, color='red', s=50, marker='x', label='Mean Point')
plt.colorbar(label='Sum of Squared Distances')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Squared Distance Sum over Grid')
plt.legend()
plt.show()
```
